```python
#importing the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#from the sample datasets of sklearn importing the breastcancer dataset

from sklearn.datasets import load_breast_cancer
```

```python
#creating an instance for the breast_cancer dataset
cancer= load_breast_cancer()
```

```python
#checking out the keyvalues of our dataset

print(cancer.keys())
```

```python
#checking out the Description value in our cancer instance of our breast_cancer dataset
print(cancer['DESCR'])
```

```python
#analysing the above description gives us much of the information on shape, size and class in our dataset

#time to use pandas DataFrame function to create our dataset in the form of a dataframe
#this dataframe will have all the data of this breastcancer dataset and the column names 
#pulled from the feature names of this sample dataset

df1= pd.DataFrame(cancer['data'], columns=cancer['feature_names'])

df1.head()
```

```python
#Lets check out the shape of our df1
df1.shape
```

```python
#we can see that we have 30 features.

#our task here is to minimize the dimensionality of our data from 30 to a less complex say 2D dataset
# as high dimensionaility leads to underfitting and  thus to inaccuracies

# The first and foremost step while performing Principal component analysis to reduce the dimensionality
#is to bring all our attributes currently under different scales to a single scale.
# this can be done using Standard Scaling or min max scaling using sklearn 

#As we can see in our data set the mean perimeter attribute 
#has altogether a different scale and span if compared to the mean compactness. 

#So, because we are planning to create a new vector space with all of these dimensions, 
#we must scale down or scale up the values, to bring their span in a close range
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

scaler.fit(df1)
scaleddata=scaler.transform(df1)
```

### scaling could have been donw using minmax scaling as well

```python
#now lets apply PCA to reduce our dimesionality

from sklearn.decomposition import PCA

pcaobject= PCA(n_components=2)
```

```python
pcaobject.fit(scaleddata)
```

```python
Xi=pcaobject.transform(scaleddata)
```

```python
Xi.shape
```

```python
scaleddata.shape
```

```python
#our data set is ready right now as Xi and has 2 columsn
print(Xi)
```

```python
# Just to quickly visualise our data set

plt.figure(figsize=(12,9))
plt.scatter(Xi[:,0], Xi[:,1], c=cancer.target)

plt.xlabel('PC-1')
plt.ylabel('PC-2')
```
